```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
titanic = pd.read_csv('./data/titanic_train.csv')
```

```python
titanic.head()
```

```python
titanic[['Age']].describe()
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False,cbar=False,cmap='viridis')
plt.show()
```

```python
# Family size feature
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch']
titanic.drop('SibSp',axis=1,inplace=True)
titanic.drop('Parch',axis=1,inplace=True)
```

```python
categorical_cols = ["Survived", "Pclass", "Sex", "FamilySize","Embarked"]
continuous_cols = ["Age", "Fare"]
```

## Countplots

```python
fig = plt.figure(figsize=(20, 10))
for i, col in enumerate(categorical_cols):
    plt.subplot(2,3, i+1)
    sns.countplot(x=col,data=titanic,palette='RdBu_r');
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
for i, col in enumerate(categorical_cols[1:]):
    plt.subplot(2,2, i+1)
    sns.countplot(x=col,hue = "Survived", data=titanic,palette='RdBu_r');
plt.show()
```

## Histograms

```python
fig = plt.figure(figsize=(15, 5))
for i, col in enumerate(continuous_cols):
    plt.subplot(1,2, i+1)
    sns.distplot(titanic[col].dropna());
plt.show()
```

```python
fig = plt.figure(figsize=(15, 10))
sub = 1
for col in continuous_cols:
    for is_survived in [0,1]:
        plt.subplot(2,2, sub)
        sns.distplot(titanic[titanic.Survived==is_survived][col].dropna(), bins=20, kde=False);
        plt.title("{} (Survived = {})".format(col, is_survived) )
        sub += 1
plt.show()
```

## Swarmplots 

```python
fig = plt.figure(figsize=(15, 10))
sub = 1
for col in continuous_cols:
    for cat_col in categorical_cols[1:]:
        plt.subplot(3,3, sub)
        sns.swarmplot(x=cat_col, y = col, hue = "Survived", data=titanic)
        sub = sub + 1
plt.show()
```

```python
titanic_df = pd.read_csv('./data/titanic_train.csv')

#Converting Sex
sex_conv = lambda x : 1 if x=='male' else 0 
titanic_df['Is_male'] = titanic_df['Sex'].apply(sex_conv)

# Adding Family column
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df.drop('SibSp',axis=1,inplace=True)
titanic_df.drop('Parch',axis=1,inplace=True)

# Adding "Title" column
titanic_df['Title']=0
for i in titanic_df:
    titanic_df['Title']=titanic_df['Name'].str.extract('([A-Za-z]+)\.', expand=False)  

# Map of aggregated titles:
titles_dict = {'Capt': 'Other','Major': 'Other', 'Jonkheer': 'Other','Don': 'Other',
               'Sir': 'Other','Dr': 'Other','Rev': 'Other','Countess': 'Other',
               'Dona': 'Other','Mme': 'Mrs','Mlle': 'Miss','Ms': 'Miss',
               'Mr': 'Mr','Mrs': 'Mrs','Miss': 'Miss', 'Master': 'Master',
               'Lady': 'Other'}
titanic_df['Title'] = titanic_df['Title'].map(titles_dict)

# Dropping some columns
titanic_df.drop(['Ticket', 'Cabin', "Name"], axis=1, inplace = True)

# Imputing Age  
titanic_df['Age'] = titanic_df.groupby('Title')['Age'].transform(lambda x: x.fillna(np.round(x.mean())))

# Converting Age
titanic_df['Age'] =pd.cut(titanic_df['Age'], bins=[1, 12, 50, 200], labels=['Child','Adult','Elder'])
```

```python
titanic_df = pd.read_csv('./data/titanic_train.csv')

#Converting Sex
sex_conv = lambda x : 1 if x=='male' else 0 
titanic_df['Is_male'] = titanic_df['Sex'].apply(sex_conv)

# Adding Family column
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df.drop('SibSp',axis=1,inplace=True)
titanic_df.drop('Parch',axis=1,inplace=True)

# Adding "Title" column
titanic_df['Title']=0
for i in titanic_df:
    titanic_df['Title']=titanic_df['Name'].str.extract('([A-Za-z]+)\.', expand=False)  
    
# Imputing Age  
titanic_df['Age'] = titanic_df.groupby('Title')['Age'].transform(lambda x: x.fillna(np.round(x.mean())))

# Map of aggregated titles:
titles_dict = {'Capt': 'Other','Major': 'Other', 'Jonkheer': 'Other','Don': 'Other',
               'Sir': 'Other','Dr': 'Other','Rev': 'Other','Countess': 'Other',
               'Dona': 'Other','Mme': 'Mrs','Mlle': 'Miss','Ms': 'Miss',
               'Mr': 'Mr','Mrs': 'Mrs','Miss': 'Miss', 'Master': 'Master',
               'Lady': 'Other'}
titanic_df['Title'] = titanic_df['Title'].map(titles_dict)

# Dropping some columns
titanic_df.drop(['Ticket', 'Cabin', "Name"], axis=1, inplace = True)

# Converting Age
titanic_df['Age'] =pd.cut(titanic_df['Age'], bins=[1, 12, 50, 200], labels=['Child','Adult','Elder'])
```
